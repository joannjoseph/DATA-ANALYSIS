```python
# QUESTION 1.  POKEMON WITH TOP 10 HIT POINTS
```

```python
# TOP POKENMON WITH HIGHEST HIT POINTS
```

```python
df.loc[df.HP>=150,['Name', 'HP']]
```

```python
# BAR GRAPH VISUALIZATION OF HIT POINTS
```

```python
x=df.Name
y =df.HP
plt.figure(figsize=(30,8))
#  Bar plot
plt.bar( x,y, color ='red',width = 1)
plt.xlabel("Name of Pokemon")
plt.ylabel("Hit points")
plt.title("HIT POINTS OF POKEMON")
plt.show()

```
